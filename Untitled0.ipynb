# Data Visualization

Scatter Plot - [ Charging Cost vs. Energy Consumed by Charger Type ]

```python
def plot_scatter_energy_vs_cost(data):

    """
    Plots a scatter plot showing the relationship between Energy Consumed (kWh)
    and Charging Cost (USD) by Charger Type.
    """

    # Setting the figure size for better readability
    plt.figure(figsize=(10, 6))

    # Extracting unique charger types to differentiate in the scatter plot
    charger_types = data['Charger Type'].unique()

    # Defining color list to assign a color to each charger type
    colors = ['blue', 'green', 'red']

    # Looping over each charger type to plot data points with specific colors
    for i, charger_type in enumerate(charger_types):
        # Filter data for the current charger type
        subset = data[data['Charger Type'] == charger_type]

        # Plot each subset with a unique color, size, and transparency
        plt.scatter(subset["Energy Consumed (kWh)"],
                    subset["Charging Cost (USD)"],
                    label=charger_type, s=100, alpha=0.7,
                    color=colors[i % len(colors)])

    # Adding plot title and labels for x and y axes
    plt.title("Charging Cost vs. Energy Consumed by Charger Type")
    plt.xlabel("Energy Consumed (kWh)")
    plt.ylabel("Charging Cost (USD)")

    # Displaying legend with the title for charger types
    plt.legend(title="Charger Type")

    # Adding a grid to improve plot readability
    plt.grid(True)

    # Display the plot
    plt.show()
```

```python
# Fuction Calling
plot_scatter_energy_vs_cost(ev_charging_df)
```

Pie Chart - Distribution of Charger Types

```python
def plot_charger_type_distribution(data):

    """
    Plots a pie chart displaying the distribution of Charger Types.
    """

    # Defining plot size
    plt.figure(figsize=(8, 8))

    # Count the occurrences of each charger type
    charger_counts = data['Charger Type'].value_counts()

    # Plot pie chart with percentage labels
    plt.pie(charger_counts, labels=charger_counts.index, autopct='%1.1f%%',
            startangle=140)

    # Adding title
    plt.title("Distribution of Charger Types")

    # Show the plot
    plt.show()
```

```python
# Fuction Calling
plot_charger_type_distribution(ev_charging_df)
```

Pie Chart - Proportion of Charging Sessions by Time of Day

```python
def plot_Charging_Sessions(data):

    """
    Plots a pie chart for proportion of Charging Sessions by Time of Day
    """

    # Plotting Pie Chart for Proportion of Charging Sessions by Time of Day
    plt.figure(figsize=(8, 8))
    time_of_day_counts = ev_charging_df['Time of Day'].value_counts()

    plt.pie(time_of_day_counts, labels=time_of_day_counts.index,
            autopct='%1.1f%%', startangle=140)

    # Adding title
    plt.title("Proportion of Charging Sessions by Time of Day")

    # Show the plot
    plt.show()
```

```python
# Fuction Calling
plot_Charging_Sessions(ev_charging_df)
```

Box Plot - Energy Consumed by User Type

```python
def plot_energy_consumption_by_user_type(data):

    """
    Plots a box plot showing Energy Consumed (kWh) by User Type.
    """

    # Defining plot size
    plt.figure(figsize=(10, 6))

    # Creating box plot to show distribution of energy consumed by user type
    sns.boxplot(data=data, x="User Type", y="Energy Consumed (kWh)", palette="Set2")

    # Adding title and labels
    plt.title("Energy Consumed by User Type")
    plt.xlabel("User Type")
    plt.ylabel("Energy Consumed (kWh)")
    plt.grid(True)

    # Show the plot
    plt.show()
```

```python
# Fuction Calling
plot_energy_consumption_by_user_type(ev_charging_df)
```

Violin Plot - Distribution of Energy Consumption by User Type in EV Charging

```python
def plot_energy_consumption_by_user_type_violin(data):

    """
    Plots a violin plot showing Distribution of Energy Consumption by User Type
    in EV Charging
    """

    # Plotting the violin plot with different colors for each user type
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='User Type', y='Energy Consumed (kWh)',
                   data=ev_charging_df, palette="Set2")

    # Adding title and labels
    plt.title("Energy Consumed by User Type")
    plt.xlabel("User Type")
    plt.ylabel("Energy Consumed (kWh)")
    plt.grid(True)

    # Show the plot
    plt.show()
```

```python
# Fuction Calling
plot_energy_consumption_by_user_type_violin(ev_charging_df)
```
